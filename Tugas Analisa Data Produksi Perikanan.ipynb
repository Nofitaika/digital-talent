# 1. Deskripsi Data
Data Produksi Perikanan merupakan data dari Kementerian Kelautan dan Perikanan yang diambil dari dataset Satu Data Indonesia.
Data ini (seperti yang tertampil di atas) terdiri dari 11 variabel, yaitu ID (id record), ProvinsiID (kode provinsi), NamaProvinsi, IkanID (id jenis ikan), NamaIkan (jenis ikan yang ditangkap), JenisPerairan (jenis perairan tempat penangkapan ikan), PeriodeID, Periode (periode pencatatan data), Tahun (tahun pencatatan data), Volume (jumlah ikan yang ditangkap dalam ton), dan Nilai (nilai jual ikan, dalam satuan rupiah).

Data produksi perikanan memiliki 76791 baris dengan 9 variabel pertama memiliki data yang lengkap, sedangkan variabel Volume dan Nilai memiliki missing data sebanyak 39399 dan 27709 data. Dataset produksi perikanan ini merupakan data yang dikumpulkan dari tahun 2000 - 2012 dari 33 provinsi di Indonesia.

# 2. Statistik Data
Dengan tanpa menghilangkan missing data akan dilakukan beberapa perhitungan statistik sederhana seperti berikut.

```python
total_volume = data_ikan.Volume.sum()
maks_volume = data_ikan.Volume.max()
total_nilai = data_ikan.Nilai.sum()
maks_nilai = data_ikan.Nilai.max()
rata_nilai = data_ikan.Nilai.mean()

print("Total penangkapan ikan dari tahun 2000-2012 sebanyak", total_volume, "ton")
print("Hasil penangkapan terbanyak yang pernah diperoleh adalah sebanyak", maks_volume, "ton")
print("Total nilai jualnya adalah", total_nilai, "rupiah")
print("Hasil penjualan terbanyaknya adalah", maks_nilai, "rupiah")
print("Rata - rata penjualan ikan dari tahun 2000-2012 adalah", rata_nilai, "rupiah")
```

# 3. Visualisasi Sederhana

```python
ax = data_ikan.groupby('Tahun').mean()['Volume']
ax = ax.plot(kind='line', title ="Rata-Rata Hasil Penangkapan Ikan (ton) Berdasarkan Tahun", fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

```python
ax = data_ikan.groupby('NamaProvinsi').mean()['Volume']
ax = ax.plot(kind='bar', title ="Rata-Rata Hasil Penangkapan Ikan (ton) Berdasarkan Provinsi", fontsize=12)
ax.set_xlabel("Provinsi", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

```python
ax = data_ikan.groupby('JenisPerairan').mean()['Volume']
ax = ax.plot(kind='bar', title ="Rata-Rata Hasil Penangkapan Ikan (ton) Berdasarkan Jenis Perairan", fontsize=12)
ax.set_xlabel("Jenis Perairan", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```
